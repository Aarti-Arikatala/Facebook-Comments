```python
#PREPROCESSING STEPS
#1. Tokenization and removal of stop words
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

df = pd.read_excel('Emoji_convert_Woodford.xlsx')
j=[]

stop_words = set(stopwords.words('english'))

for i in df['body_text']:
    #removing punctuation
    i = re.sub(r'[^\w\s]','',i)
    i=i.lower()
    #tokenization
    words = nltk.tokenize.word_tokenize(i)
    #removing stop words
    words_list = [w for w in words if not w in stop_words]
    words_list = str(words_list)[1:-1]
    words_list = words_list.replace('"',"")
  
    j.append(words_list)
df.loc[:, 'body_text'] = j
#Returns a file where all the comments are broken into tokens and the stop words are moved
df.to_excel("Emoji_token_Woodford.xlsx", encoding = 'utf-8')
```
